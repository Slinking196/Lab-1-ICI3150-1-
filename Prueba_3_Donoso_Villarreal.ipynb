## Hipótesis 1
### 1° Hipótesis: En comunas urbanas hay más defunciones por obesidad que en comunas rurales.

####Gráfico de barras SIN considerar la diferencia entre la cantidad de comunas urbanas con las rurales.

````python
df_hipotesis_1 = df_hipotesis

(
  ggplot(df_hipotesis_1)+
  aes(x='Clasificación', y='..count..', fill = "Clasificación")+
  geom_bar()+
  theme_grey()+
  ggtitle("Concurrencias de Defunciones por Obesidad [Urbano/Rural]")+
  xlab("TIPO DE COMUNA")+
  ylab("DEFUNCIONES")+
  scale_fill_brewer(name="Tipo de comuna",palette=14)
)
````

####Gráfico de barra considerando la diferencia de población total entre la zona rural y la zona urbana.

````python
#Conseguir población total de habitantes que viven en zona rural o urbana
poblacion = pd.DataFrame()
poblacion = df_Censo2017
poblacion = poblacion.drop(['Cod_reg','Región','cod_com', 'Comuna','Km.2','Densidad','Tipo_com'], axis = 1)
poblacion_final = poblacion.groupby(['Clasificación']).sum()
poblacion_final
````

````python
#Conseguir total de casos de defunciones por obesidad en zona urbana o rural
total_casos = pd.DataFrame()
total_casos = df_hipotesis
total_casos = total_casos.drop([	"Sexo",	"Edad",	"Cod_comuna",	"Comuna",	"Región",	"Cod_Causa_Simple",	"Diagnóstico_Simple",	"Lugar defunción",	"Tipo_com"] , axis = 1)
total_casos.columns = ["Defunciones", "Clasificación"]
total_casos_final = total_casos.groupby(['Clasificación']).count()
total_casos_final
````

````python
#Merge entre los 2 dataframes anteriores
df_pob_def = pd.merge(poblacion_final, total_casos_final, left_on='Clasificación', right_on='Clasificación', how = 'left')
df_pob_def = df_pob_def.drop( ["Mixta"] ) #Se eliminan las comunas mixtas del estudio
df_pob_def
````

````python
#Se calcula con regla simple de 3 la relación entre la población y la defunción por obesidad
df_pob_def['Defunciones/1.000.000 Habitantes'] = [ (91*1000000)/2241045 , (467*1000000)/13095911 ]
df_pob_def.reset_index(inplace=True,drop=False)
df_pob_def
````

````python
(
  ggplot(df_pob_def)+
  aes(x='Clasificación', y='Defunciones/1.000.000 Habitantes', fill = 'Clasificación')+
  geom_bar(stat="identity")+
  theme_grey()+
  ggtitle("Concurrencias de Defunciones por Obesidad [Urbano/Rural]")+
  xlab("TIPO DE COMUNA")+
  ylab("Defunciones/1.000.000 Habitantes")
)
````

````python
#.- Obtener promedio de distancia de cada región hasta Santiago Centro
promedioDistanciaXRegion = pd.DataFrame()
promedioDistanciaXRegion = conteoXcomunaYdistancia
promedioDistanciaXRegion = promedioDistanciaXRegion.drop(['frecuencia','Nombre comuna'], axis = 1)
promedioDistanciaXRegion_Final = promedioDistanciaXRegion.groupby(['Región']).mean()
promedioDistanciaXRegion_Final.reset_index(inplace=True,drop=False)
promedioDistanciaXRegion_Final['Distancia Hasta Santiago(km)'] = promedioDistanciaXRegion_Final['Distancia Hasta Santiago(km)'].astype(int)
promedioDistanciaXRegion_Final.head()
````
